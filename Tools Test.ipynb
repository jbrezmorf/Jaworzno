## Test of column converters for the pandas read_table function.

```python
from jaworzno_tools import *
    
print(coord_converter('19.25029°E')) 
print(coord_converter('19.25082°E')) 
 
print(coord_converter('19°13´45.132"E'))    
print(coord_converter('50.°11´21.880"N'))

def float_converter(x):
    """
    Convert values like '<0.01' to the value.
    """
    try:
        if x[0] == '<':
            x = x[1:]
        return float(x)
    except:
        return 0.0
print(float_converter('12.86'))
```
